```python
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
```

```python
df.shape
```

```python
x_train , x_test , y_train , y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_train
```

```python
x_train_tensor = torch.from_numpy(x_train)
x_test_tensor = torch.from_numpy(x_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
y_train_tensor.shape
```

```python
class MYSimpleNN( ):
  def __init__(self,x):
    self.weights =torch.rand(x.shape[1],1,dtype=torch.float64,requires_grad=True)
    self.bias = torch.zeros(1,dtype=torch.float64 , requires_grad=True)

  def forward(self,x):
    z=torch.matmul(x,self.weights)+self.bias
    y_hat = torch.sigmoid(z)
    return y_hat

  def binary_loss(self , y_hat , y):
    # to avoid log(0) = infinity
    epsilone = 1e-7
    y_hat = torch.clamp(y_hat,epsilone,1-epsilone)

    loss = (-y*torch.log(y_hat)-(1-y)*torch.log(1-y_hat)).mean()
    return loss
```

```python
lr=0.1
epochs = 50
```

```python
model = MYSimpleNN(x_train_tensor)
# model.weights
```

```python
# model.bias
```

```python
for epoch in range(epochs):
  y_hat=model.forward(x_train_tensor)
  # print(y_hat)

  # loss
  loss = model.binary_loss(y_hat,y_train_tensor)
  # print(f'epoch: {epoch+1} , loss: {loss.item()}')

  loss.backward()

  # parameter update
  with torch.no_grad():
    model.weights -= lr*model.weights.grad
    model.bias -= lr*model.bias.grad

  # zero gradient
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  # print loss in each epoch
  print(f'epoch: {epoch+1} , loss: {loss.item()}')
```

```python
with torch.no_grad():
 y_hat= model.forward(x_test_tensor)
 y_hat =(y_hat>0.5).float()

accuracy = (y_hat==y_test_tensor).float().mean()
print(f'accuracy: {accuracy.item()}')
```
