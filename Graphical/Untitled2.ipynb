```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
y = [20, 25, 35, 55]
x = [10, 20, 30, 40]
```

```python
plt.plot(x,y)
plt.title('Linear Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# plt.xlim([10,60])
plt.ylim([0,60])
```

```python
import numpy as np
```

```python
x = np.array([1,2,5,4,8])
fig = plt.figure(figsize=(5,5))

plt.plot(x,x**2, label='Square', marker='*')
plt.plot(x,x**3, label='Cube', marker='o')
plt.xlim(1,20)
plt.ylim(1,600)
plt.legend()
plt.show()
```
